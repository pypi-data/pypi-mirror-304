Ревизия свойств воды

чтобы посмотреть, что и как считается

```python
import matplotlib.pyplot as plt
import numpy as np


import neftpy.upvt_wat as pvtw
import neftpy.upvt_np_vect as pvt
```

Объемный коэффициент воды

```python
t_list = np.linspace(273, 350, 100)
p = 1

plt.plot(t_list, pvtw.unf_bw_McCain_m3m3(t_K=t_list, p_MPaa=0.1))
plt.plot(t_list, pvtw.unf_bw_McCain_m3m3(t_K=t_list, p_MPaa=1))
plt.plot(t_list, pvtw.unf_bw_McCain_m3m3(t_K=t_list, p_MPaa=10))
plt.title('объемный коэффициент воды от температуры')
```

```python
p_list = np.linspace(0.1, 20, 100)

plt.plot(p_list, pvtw.unf_bw_McCain_m3m3(t_K=278, p_MPaa=p_list))
plt.plot(p_list, pvtw.unf_bw_McCain_m3m3(t_K=300, p_MPaa=p_list))
plt.plot(p_list, pvtw.unf_bw_McCain_m3m3(t_K=350, p_MPaa=p_list))
plt.title('объемный коэффициент воды от давления')
```

# связь плотности воды и солености

```python
gw_list = np.linspace(1, 1.3, 100)

plt.plot(gw_list, pvtw.unf_salinity_from_gamma_water_ppm(gw_list))

sal_list = np.linspace(1, 350000, 100)
plt.plot(pvtw.unf_gamma_water_from_salinity_m3m3(sal_list),sal_list )

plt.title("связь плотности воды и солености")
```

# вязкость воды

```python
t_K = 300
p_MPa = 1
sal_list = np.linspace(1, 350000, 1000)
plt.plot(pvtw.unf_gamma_water_from_salinity_m3m3(sal_list), pvtw.unf_mu_water_McCain_cP(t_K, p_MPa, sal_list) )
plt.plot(pvtw.unf_gamma_water_from_salinity_m3m3(sal_list), pvtw.unf_mu_water_McCain_cP(t_K+30, p_MPa, sal_list) )
plt.plot(pvtw.unf_gamma_water_from_salinity_m3m3(sal_list), pvtw.unf_mu_water_McCain_cP(t_K+60, p_MPa, sal_list) )

plt.title("связь вязкости и плотности")
```
